```python
folder = '../class_data/Apple-20220324T003432Z-001'
images = [os.path.join(root, filename)
          for root, dirs, files in os.walk(folder)
          for filename in files
          if filename.lower().endswith('.jpg')]

ic = []
for img in images:
    ic.append(io.imread(img))
```
