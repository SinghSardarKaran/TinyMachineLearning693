```python
import os
MODELS_DIR = 'models/'
if not os.path.exists(MODELS_DIR):
    os.mkdir(MODELS_DIR)
MODEL_TF = MODELS_DIR + 'model'
MODEL_NO_QUANT_TFLITE = MODELS_DIR + 'model_no_quant.tflite'
MODEL_TFLITE = MODELS_DIR + 'model.tflite'
MODEL_TFLITE_MICRO = MODELS_DIR + 'model.cc'
import tensorflow as tf
##!pip install tensorflow
import tensorflow as tenf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
seed = 1
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# TensorFlow is an open source machine learning library
import tensorflow as tf

# Keras is TensorFlow's high-level API for deep learning
from tensorflow import keras
# Numpy is a math library
import numpy as np
# Pandas is a data manipulation library 
import pandas as pd
# Matplotlib is a graphing library
import matplotlib.pyplot as plt
# Math is Python's math library
import math

# Set seed for experiment reproducibility
seed = 1
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# We'll use Keras to create a simple model architecture
model_1 = tf.keras.Sequential()

model_1.add(keras.layers.Dense(128, activation ='elu' ,input_shape=(a_train.shape[1],)))

# Final layer is 3 neurons, since we want to output into 3 classes
model_1.add(keras.layers.Dense(2, activation='softmax'))

# Compile the model using the standard 'adam' optimizer and Sparse categorical crossentropy
# Compile the model
model_1.compile(loss=tf.keras.losses.sparse_categorical_crossentropy,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])
print(model_1.summary())
```

```python
model_1.save(MODEL_TF)
history_1 = model_1.fit(a_train, b_train, epochs=20, batch_size=10)
```

```python
train_pred = model_1.predict(a_train)
train_pred
```

```python
train_pred = np.argmax(train_pred,axis=1)
train_pred
```

```python
from sklearn.metrics import classification_report

print(classification_report(b_train,train_pred))
```

```python
test_pred = model_1.predict(a_test)
train_pred
```

```python
test_pred = np.argmax(test_pred,axis=1)
test_pred
```

```python
print(classification_report(b_test,test_pred))
```

```python
# Confusion Matrix

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

conf_mat = confusion_matrix(b_test,test_pred)
class_label = ["wave","fistbump"]
df = pd.DataFrame(conf_mat, index = class_label, columns = class_label)
sns.heatmap(df, annot = True,fmt="d")
plt.title(" Radom Forest Wave-Fistbump train data")
plt.xlabel("Pred Label")
plt.ylabel("T Label")
plt.show()

```

```python
# Convert the model to the TensorFlow Lite format without quantization
converter = tf.lite.TFLiteConverter.from_saved_model(MODEL_TF)
model_no_quant_tflite = converter.convert()

# Save the model to disk
open(MODEL_NO_QUANT_TFLITE, "wb").write(model_no_quant_tflite)


# Convert the model to TensorFlow Lite format with quantization
converter = tf.lite.TFLiteConverter.from_saved_model(MODEL_TF)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
model_tflite = converter.convert()
#saving converted model in "converted_quant_model.tflite" file
open(MODEL_TFLITE, "wb").write(model_tflite)
```

```python
# Calculate size
size_tf = os.path.getsize(MODEL_TF)
size_no_quant_tflite = os.path.getsize(MODEL_NO_QUANT_TFLITE)
size_tflite = os.path.getsize(MODEL_TFLITE)
```

```python
# Compare size
pd.DataFrame.from_records(
    [["TensorFlow", f"{size_tf} bytes", ""],
     ["TensorFlow Lite", f"{size_no_quant_tflite} bytes ", f"(reduced by {size_tf - size_no_quant_tflite} bytes)"],
     ["TensorFlow Lite Quantized", f"{size_tflite} bytes", f"(reduced by {size_no_quant_tflite - size_tflite} bytes)"]],
     columns = ["Model", "Size", ""], index="Model")
```
