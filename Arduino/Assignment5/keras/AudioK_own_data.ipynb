```python
down1 = down.iloc[:,:64]
up1 = up.iloc[:,:64]
go1 = go.iloc[:,:64]
stop1 = stop.iloc[:,:64]
```

```python
r = r1 + r2 + r3 + r4
r = np.array(r)
r
```
