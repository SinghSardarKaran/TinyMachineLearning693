```python
# 'stop
import pandas as pd
import glob

folder_path = r'C:\Users\sarda\GitHub\TinyMachineLearning693\Arduino\Assignment3\data\Stop'
file_list = glob.glob(folder_path + "/*.csv")
st_dataframe = pd.DataFrame(pd.read_csv(file_list[0]))

for i in range(1,len(file_list)):
    data = pd.read_csv(file_list[i])
    df = pd.DataFrame(data)
    st_dataframe = pd.concat([st_dataframe,df],axis=0)
print(st_dataframe)
```

```python
u1 = up_dataframe.iloc[:,:-1]
dwn1 = down_dataframe.iloc[:,:-1]
go1 = go_dataframe.iloc[:,:-1]
st1 = st_dataframe.iloc[:,:-1]
```

```python
r1 = np.concatenate([dwn1,u1,go1,st1])
r1.shape

```

```python
q1 = np.array([0]*475 + [1] * 475 + [2]*475 + [3] * 475)
q1

```

```python
print(q1.shape)

a_train,a_test,b_train,b_test = train_test_split(r1,q1,train_size = 0.8,random_state = 0)
```

```python
## logictic regression 
```

## Random forest

```python
rdf = RandomForestClassifier(max_depth=2, random_state=0)


rdf.fit(a_train, b_train)

```

```python
rf1 = rdf.predict(a_test)
rf1

```

```python
print(classification_report(b_test,rf1))


#Confusion matrix
```

```python
cf_matrix5 = confusion_matrix(b_test, rf1)

class_label = ["down","up","go","stop"]
df5 = pd.DataFrame(cf_matrix5, index = class_label, columns = class_label)
sns.heatmap(df5, annot = True,fmt="d")
```
