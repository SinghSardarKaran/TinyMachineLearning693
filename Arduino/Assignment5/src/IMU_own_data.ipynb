```python
import sys,os
import pandas as pd
import numpy as np
```

## lets First read the wave data

```python
waved = pd.read_csv("../data/wave_data.csv")
waved
```

## lets now drop the null values from the wave data

```python
waved.dropna()
waved
```

## reading the main values of wave data

```python
waved = waved.iloc[:,:6]
waved[:6]
```

```python
wavedd = waved.values.tolist()
wavedd[:6]
```

## coverting them to resultant format

```python
n = 119
wve = [wavedd[i:i + n] for i in range(0, len(wavedd), n)]
print(wve[:2])
```

**final value**

```python
W = np.zeros(50,).astype(int)
W.shape
```

```python
W
```

```python
## lets read fist bump data 
```

```python
fstbmp = pd.read_csv("../data/Fistbump.csv")
fstbmp
```

```python
fstbmp = fstbmp.iloc[:,:6]
fstbmp[:6]
```

```python
fsB = fstbmp.values.tolist()
fsB[:6]
```

```python
n = 119
fsBB = [fsB[i:i + n] for i in range(0, len(fsB), n)]
print(fsBB[:2])
```

```python
np.array(fsBB).shape
```

```python
f = np.ones(50,)
f.shape
```

```python
## combining two varibles
```

```python
combination_wf1 = np.concatenate([wve,fsBB])
combination_wf1.shape
```

```python
## combining target variables
```

```python
combination_wf2 = np.concatenate([W,f])
combination_wf2.shape
```

```python
## reshape check
```

```python
combination_wf1_reshaped = combination_wf1.reshape(100,119*6)
combination_wf1_reshaped.shape
```

```python
from sklearn.model_selection import train_test_split
a_train, a_test, b_train, b_test = train_test_split(combination_wf1_reshaped,combination_wf2,train_size=0.8)
print(a_train.shape)
print(a_test.shape)
print(b_train.shape)
print(b_test.shape)
```

## random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
from micromlgen import port
from sklearn.metrics import classification_report

##!pip install micromlgen

```

```python
rfm = RandomForestClassifier(max_depth=2)
rfm.fit(a_train,b_train)
```

```python
r_test_rf = rfm.predict(a_test)
r_test_rf
```

```python
r_train_rf = rfm.predict(a_train)
r_train_rf
```

```python
print(classification_report(b_train,r_train_rf))
```

```python
print(classification_report(b_test,r_test_rf))
```

```python
# Confusion Matrix

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

conf_mat = confusion_matrix(b_train,r_train_rf)
class_label = ["wave","fistbump"]
df = pd.DataFrame(conf_mat, index = class_label, columns = class_label)
sns.heatmap(df, annot = True,fmt="d")
plt.title(" Radom Forest Wave-Fistbump train data")
plt.xlabel("Pred Label")
plt.ylabel("T Label")
plt.show()

```

```python
# Confusion Matrix



conf_mat = confusion_matrix(b_test,r_test_rf)
class_label = ["wave","fistbump"]
df = pd.DataFrame(conf_mat, index = class_label, columns = class_label)
sns.heatmap(df, annot = True,fmt="d")
plt.title(" Radom Forest Wave-Fistbump train data")
plt.xlabel("Pred Label")
plt.ylabel("T Label")
plt.show()

```

```python
with open("../models/Iris_Classifier.h", "w") as file:
    file.write(port(rfm, classname="Classifier"))
```
